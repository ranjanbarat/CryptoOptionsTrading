```python
for i in range(len(order_dataframe["expiry"])):
    print(f"{i} : {order_dataframe.loc[i]["expiry"]}")
    if expiry["daily"]["current"]["date"] == order_dataframe.loc[i]["expiry"]:
        print(f"Its Matching the Current_Daily Date: {order_dataframe.loc[i]["expiry"]}")
    elif expiry["daily"]["next"]["date"] == order_dataframe.loc[i]["expiry"]:
        print(f"Its Matching the Next_Daily Date: {order_dataframe.loc[i]["expiry"]}")
    elif expiry["daily"]["next_to_next"]["date"] == order_dataframe.loc[i]["expiry"]:
        print(f"Its Matching the Next_to_Next_daily Date: {order_dataframe.loc[i]["expiry"]}")
        
    elif expiry["weekly"]["current"]["date"] == order_dataframe.loc[i]["expiry"]:
        print(f"Its Matching the Current_Weekly Date: {order_dataframe.loc[i]["expiry"]}")
    elif expiry["weekly"]["next"]["date"] == order_dataframe.loc[i]["expiry"]:
        print(f"Its Matching the Next_Weekly Date: {order_dataframe.loc[i]["expiry"]}")
    elif expiry["weekly"]["next_to_next"]["date"] == order_dataframe.loc[i]["expiry"]:
        print(f"Its Matching the Next_to_Next_weekly Date: {order_dataframe.loc[i]["expiry"]}")

    elif expiry["monthly"]["current"]["date"] == order_dataframe.loc[i]["expiry"]:
        print(f"Its Matching the Current_Monthly Date: {order_dataframe.loc[i]["expiry"]}")
    elif expiry["monthly"]["next"]["date"] == order_dataframe.loc[i]["expiry"]:
        print(f"Its Matching the Next_Monthly Date: {order_dataframe.loc[i]["expiry"]}")
    elif expiry["monthly"]["next_to_next"]["date"] == order_dataframe.loc[i]["expiry"]:
        print(f"Its Matching the Next_to_Next_Monthly Date: {order_dataframe.loc[i]["expiry"]}")

    elif expiry["quarterly"]["current"]["date"] == order_dataframe.loc[i]["expiry"]:
        expiry["quarterly"]["current"]["positional_delta"] = order_dataframe.loc[i]["positional_delta"]
        expiry["quarterly"]["current"]["position_exist"] = True
        
        print(f"Its Matching the Current_Quarterly Date: {order_dataframe.loc[i]["expiry"]}, ", 
              f"Position_Exist : {expiry["quarterly"]["current"]["position_exist"]}, ",
              f"Positional_Delta is : {expiry["quarterly"]["current"]["positional_delta"]}")
    elif expiry["quarterly"]["next"]["date"] == order_dataframe.loc[i]["expiry"]:
        print(f"Its Matching the Next_Quarterly Date: {order_dataframe.loc[i]["expiry"]}")
    elif expiry["quarterly"]["next_to_next"]["date"] == order_dataframe.loc[i]["expiry"]:
        print(f"Its Matching the Next_to_Next_Quarterly Date: {order_dataframe.loc[i]["expiry"]}")
    


        
```
